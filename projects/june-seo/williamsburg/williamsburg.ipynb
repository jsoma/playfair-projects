```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
df = pd.read_csv("rental_inventory_time_series_prices_All_Types_One_Bedroom.csv")
```

```python
df
```

```python
# Neighborhood level
brooklyn_df = df[df['Area']=='Brooklyn']
neighborhood_df = df[df['AreaType']=='Neighborhood']
```

```python
bk_df = neighborhood_df.append(brooklyn_df)
```

```python
bk_df = bk_df.transpose()
```

```python
bk_df
```

```python
new_header = bk_df.iloc[0]
bk_df = bk_df[3:]
bk_df = bk_df.rename(columns=new_header)
```

```python
nine_df = bk_df[:10] #2009
ten_df = bk_df[10:22] #2010
eleven_df = bk_df[22:34] #2011
twelve_df = bk_df[34:46] #2012
thirteen_df = bk_df[46:58] #2013
fourteen_df = bk_df[58:70] #2014
fifteen_df = bk_df[70:82] #2015
sixteen_df = bk_df[82:]#2016
```

```python
bk_df[['Williamsburg', 'Brooklyn']].plot()
```

```python
fig, ax = plt.subplots(figsize=(30,5))

ax = bk_df[['Williamsburg','Brooklyn']].plot(ax=ax)

ax.set_ylim(1500,3500)
ax.set_xlim()
ax.tick_params(
    which='major',
    bottom='on',
    left='off',
    right='off',
    top='off')

ax.grid(color='grey', linestyle=':', linewidth=0.5)
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("longlongline_rent.pdf", transparent=True)
```
