# subway entrance and exit data 

```python
subway_df=pd.read_csv('StationEntrances.csv')
```

```python
subway_df.columns
```

# matching subway locations to citibike locations

```python
subway_df['latitude_f']= subway_df['Latitude'].apply(lambda x: "{0:.3f}".format(x))
subway_df['longitude_f'] = subway_df['Longitude'].apply(lambda x: "{0:.3f}".format(x))
citibike_df['start_latitude_f'] = citibike_df['start station latitude'].apply(lambda x: "{0:.3f}".format(x))
citibike_df['start_longitude_f'] = citibike_df['start station longitude'].apply(lambda x: "{0:.3f}".format(x))
citibike_df['end_latitude_f'] = citibike_df['end station latitude'].apply(lambda x: "{0:.3f}".format(x))
citibike_df['end_longitude_f'] = citibike_df['end station longitude'].apply(lambda x: "{0:.3f}".format(x))
```

```python
subway_df['location'] =  subway_df[['latitude_f', 'longitude_f']].apply(tuple, axis=1)
```

```python
citibike_df['start_location'] =  citibike_df[['start_latitude_f', 'start_longitude_f']].apply(tuple, axis=1)
citibike_df['end_location'] = citibike_df[['end_latitude_f', 'end_longitude_f']].apply(tuple, axis=1)
```

```python
citibike_df.head()
```

### select a sample of citibike data

```python
cb_sample = citibike_df.sample(n=100000)
cb_sample.to_csv('citibike_sample.csv')
```

### make one list of citibike trips start and end locations, sort

```python
cb_starts = cb_sample['start_location'].tolist()
```

```python
cb_ends = cb_sample['end_location'].tolist()
```

```python
cb_starts_ends = cb_starts + cb_ends
```

```python
import operator
```

```python
cb_starts_ends.sort(key = operator.itemgetter(0, 1))
```

### make list of subway entrances/exits, sort

```python
subway_entrances = subway_df['location'].tolist()
```

```python
subway_entrances.sort(key = operator.itemgetter(0, 1))
```

### check if citibike location matches any subway location

```python
for item in cb_starts_ends: 
    if item in subway_entrances: 
        cb_sample['near subway'] = True
    else: 
        cb_sample['near subway'] = False
```

### split citibike sample into two data sets: one that only contain starts, the other that only contains ends

```python
cb_sample.columns
```

# find the most common trips for subscribers and customers
- make tuples of starts and ends
- groupby pairs, and make dataframe with size of groups

### make tuples of start and end stations; divide sample into subscribers and customers

```python
cb_sample['route'] = cb_sample[['start station name', 'end station name']].apply(tuple, axis=1)
```

```python
cb_sample_subscriber = cb_sample[cb_sample['usertype']=='Subscriber']
# cb_sample_subscriber.to_csv('sub_sample.csv')
```

```python
cb_sample_customer = cb_sample[cb_sample['usertype']=='Customer']
# cb_sample_customer.to_csv('cust_sample.csv')
```

### look at most popular routes, start stations, and end stations

#### subscribers

```python
#routes

#groups subscriber citibike sample by route, counts occurences of routes, returns data frame
sub_g1 = cb_sample_subscriber.groupby(['route']).size().reset_index(name="count")

#merges route frequencies with individual rows in subscriber citibike data set
result = pd.merge(cb_sample_subscriber, sub_g1, how='left', on='route')

#drops duplicate rows to create data set with one row per route with the frequency of that route
sub_g1_geo = result.drop_duplicates(subset='route')
sub_g1_geo.drop(['start station id','end station id','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location'], axis=1, inplace=True)
sub_g1_geo.columns
```

```python
#starts 

sub_g2 = cb_sample_subscriber.groupby(['start station name']).size().reset_index(name="count")
result = pd.merge(cb_sample_subscriber, sub_g2, how='left', on='start station name')
sub_g2 = result.drop(['end station id', 'end station name','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'end_latitude_f', 'end_longitude_f'], axis=1)
sub_g2_geo = result.drop_duplicates(subset='start station name')
sub_g2_geo.drop(['end station id', 'end station name','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'end_latitude_f', 'end_longitude_f'], axis=1, inplace=True)
# sub_g2_geo.sort_values(by='count', ascending=False).head(10)
```

```python
#ends 

sub_g3 = cb_sample_subscriber.groupby(['end station name']).size().reset_index(name="count")
result = pd.merge(cb_sample_subscriber, sub_g3, how='left', on='end station name')
sub_g3 = result.drop(['start station id', 'start station name','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'start_latitude_f', 'start_longitude_f'], axis=1)
sub_g3_geo = result.drop_duplicates(subset='end station name')
sub_g3_geo.drop(['start station id', 'start station name','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'start_latitude_f', 'start_longitude_f'], axis=1, inplace=True)
sub_g3_geo.columns
```

```python
frames=[sub_g2, sub_g3]
```

```python
result3 = pd.concat(frames)
```

```python
result3_s = result3.sort_values(by=['bikeid','starttime', 'stoptime'])
```

```python
result3_s['order']=result3_s['end station id'].isnull().apply(lambda x: 1 if x == True else 2)
```

```python
sub_g1_geo['route number']=pd.factorize(sub_g1_geo['route'])[0]
```

```python
result3_s.columns
```

```python
sub_g1_geo.drop(['starttime', 'stoptime', 'start station name', 'end station name',
       'bikeid', 'tripduration_min', 'start_latitude_f', 'start_longitude_f',
       'end_latitude_f', 'end_longitude_f', 'near subway','count'], axis=1, inplace=True)
sub_g1_geo['route number'].head()
```

```python
result4 = pd.merge(result3_s, sub_g1_geo, how='left', on='route')
```

```python
result4.loc[result4['end station id'].isnull(),'end station id'] = result4['start station id']
result4.loc[result4['end station name'].isnull(),'end station name'] = result4['start station name']
result4.loc[result4['end_latitude_f'].isnull(),'end_latitude_f'] = result4['start_latitude_f']
result4.loc[result4['end_longitude_f'].isnull(),'end_longitude_f'] = result4['start_longitude_f']
result4.drop(['start station id', 'start station name', 'start_latitude_f', 'start_longitude_f'], axis=1, inplace=True) 
```

```python
result4.rename(columns={'end station id': 'station id', 'end station name': 'station name', 'end_latitude_f':'latitude', 'end_longitude_f':'longitude', 'route':'route name', 'route number':'route'}, inplace=True)
result4.head(10)
```

#### customers

```python
#routes 

cust_g1 = cb_sample_customer.groupby(['route']).size().reset_index(name="count")
result = pd.merge(cb_sample_customer, cust_g1, how='left', on='route')
cust_g1_geo = result.drop_duplicates(subset='route')
cust_g1_geo.drop(['start station id','end station id','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location'], axis=1, inplace=True)
```

```python
#starts 

cust_g2 = cb_sample_customer.groupby(['start station name']).size().reset_index(name="count")
result = pd.merge(cb_sample_customer, sub_g2, how='left', on='start station name')
cust_g2_geo = result.drop_duplicates(subset='start station name')
cust_g2_geo.drop(['end station id', 'end station name', 'usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'end_latitude_f', 'end_longitude_f'], axis=1, inplace=True)
cust_g2_geo.columns
```

```python
#ends 

cust_g3 = cb_sample_customer.groupby(['end station name']).size().reset_index(name="count")
result = pd.merge(cb_sample_customer, sub_g3, how='left', on='end station name')
cust_g3_geo = result.drop_duplicates(subset='end station name')
cust_g3_geo.drop(['start station id', 'start station name','usertype','birth year','gender','tripduration', 'start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'start_location', 'end_location', 'start_latitude_f', 'start_longitude_f'], axis=1, inplace=True)
cust_g3_geo.columns
```
