```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
nyc_df = pd.read_excel("NYC-accessible-copy.xls")
```

```python
nyc_df.head()
```

```python
del nyc_df['Station']
```

```python
del nyc_df['Line']
```

```python
del nyc_df['Phase']
```

```python
nyc_df.head()
```

```python
nyc_df_sum = nyc_df.sum()
```

```python
new_nyc_df = nyc_df_sum.to_frame()
```

```python
new_nyc_df['No of accessible stations per line'] = new_nyc_df
```

```python
new_nyc_df
```

```python
new_nyc_df['No of stations per line'] = [38, 61, 34, 54, 45, 38, 22, 49, 37, 40, 41, 32, 45, 21, 30, 24, 36, 45, 39, 45, 11, 20, 22]
```

```python
new_nyc_df
```

```python
new_nyc_df['% of accessible stations per line'] = round((new_nyc_df['No of accessible stations per line']/new_nyc_df['No of stations per line'])*100)
```

```python
new_nyc_df['% of not accessible stations per line'] = 100 - new_nyc_df['% of accessible stations per line']
```

```python
new_nyc_df
```

```python
fig, ax = plt.subplots()

fig.set_facecolor('gainsboro')
ax.set_axis_bgcolor('gainsboro')

new_nyc_df.plot(kind='bar', y=['% of not accessible stations per line', '% of accessible stations per line'], stacked=True, rot=1, ax=ax, legend=True, width=0.80, figsize=(10,5))

ax.yaxis.grid(which='major', color='grey', linestyle='-', linewidth=0.2)

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tick_params(
    which='major',
    top='off', 
    left='off',
    right='off',
    bottom='off',
    labelright='off', 
    labeltop='off',
    labelbottom='on')

ax.set_title("How accesible are NYC's subway stations?")
ax.set_xlabel("Subway line")
ax.set_ylabel("% of accessible stations")

ax.set_ylim(0,100)

plt.savefig("NYC-accessible-stations.pdf")
```
