I'm using this site http://research.domaintools.com/statistics/tld-counts/ to scrap the amout of Top Level occurrences and creat a .csv with the most recent approximation of top level domains.

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = "http://www.iana.org/domains/root/db"
response = requests.get(url)
iana_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
iana = iana_soup.find_all('tr')
```

```python
Types = []
for x in iana:
    try:
        Domain = x.find('td', text = re.compile('generic|country-code')).string
    except:
        None
    Types.append(Domain)
len(Types)
```

```python
Sponsoren = []
for x in iana:
    try:
        Sponsor = x.find('td', text = re.compile('[A-Z]+|aigo+|nic.at+|.au Domain+\
        |fTLD+|.CLUB+|.CO+|iSelect+|dot Loan+|iSelect+')).string
    except:
        None
    Sponsoren.append(Sponsor)
len(Sponsoren)
```

```python
iana_list = []
for x,y,z in zip(Domains, Types, Sponsoren):
    
    iana_dict = {'Domain': x,
                'Type': y,
                'Sponsor': z}
    
    iana_list.append(iana_dict)

print(iana_list[0])
```

```python
html_str = open('data/view-source_research.domaintools.com_statistics_tld-counts_.htm', 'r').read()
```

```python
html_str = BeautifulSoup(html_str, "html.parser")
```

```python
domains = html_str.find_all('div', {'class': 'stats-container'})
```

```python
import re
index = 0
domain_list = []
for domain in domains:
    all_domain_names = domain.find_all('td', {'class': 'name'})
    for item in all_domain_names:
        item = item.string
        item = item.replace(r"\(\.+\)$", '')
        index = index + 1
        domain_dict = {'index': index, 'name': item}
        #print(item)
        domain_list.append(domain_dict)
```

```python
domain_count_list = []
index = 0
for count in domains:
    all_domain_counts = domain.find_all('td', {'class': 'amount'})
    for item in all_domain_counts:
        item = item.string
        item = item.replace(",", "")
        if item == "N/A":
            item = '0'
        item = int(item)
        index = index + 1
        count_dict = {'index': index, 'count': item}
        domain_count_list.append(count_dict)
```
